```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split


number_of_files = 16
for n in range(number_of_files):

  file = np.loadtxt("/content/drive/MyDrive/Python_MIW/PRO4/Dane/dane{}.txt".format(n + 1))

  X = file[:, [0]]
  y = file[:, [1]]

  print('Dataset:', str(n + 1))
  print()

  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # train and test sets

  # liniowy model y = ax + 1
  linear_matrix = np.hstack([X_train, np.ones(X_train.shape)]) # macierz [x, 1]
  v_linear = np.linalg.pinv(linear_matrix) @ y_train 
  linear_model = v_linear[0] * X + v_linear[1]
  print('Linear model: \n', v_linear)

  e_linear_train = sum((y_train - (v_linear[0] * X_train + v_linear[1])) ** 2) / len(X_train) 
  print('Linear model (train): \n', e_linear_train)

  e_linear_test = sum((y_test - (v_linear[0] * X_test + v_linear[1])) ** 2) / len(X_test) 
  print('Linear model (test): \n', e_linear_test)
  print('\n')

  # y = x^2 - 2*x + 1
  other_matrix = np.hstack([pow(X_train, 2), -2*pow(X_train, 1), np.ones(X_train.shape)]) # macierz [x, 1]
  v_other = np.linalg.pinv(other_matrix) @ y_train 
  other_model = v_other[0] * pow(X, 2) + (-2) * v_other[1] * pow(X, 1) + 1
  print('Other model: \n', v_other)

  e_other_train = sum((y_train - (v_other[0] * pow(X_train, 2) + (-2) * v_other[1] * pow(X_train, 1) + 1)) ** 2) / len(X_train) 
  print('Other model (train): \n', e_other_train)

  e_other_test = sum((y_test - (v_other[0] * pow(X_test, 2) + (-2) * v_other[1] * pow(X_test, 1) + 1)) ** 2) / len(X_test) 
  print('Other model (test): \n', e_other_test)
  print('\n')

  # wykresy
  plt.plot(X_test, y_test, 'bo') # test data
  plt.plot(X_train, y_train, 'ro') # train data
  plt.plot(X, linear_model) # linear model
  plt.plot(X, other_model) # other model
  plt.show()
```
